### Main Application - Login and registration Page

```python
import tkinter as tk
from tkinter import messagebox

class MainApplication(tk.Tk):
    def __init__(self):
        super().__init__()
        
        # Configure window
        self.title("Registration Page")
        self.geometry("800x800")  # Set window size
        self.configure(bg="lightblue")  # Set background color

        self.title("Login and Registration System")

        self.main_frame = tk.Frame(self)
        self.main_frame.pack()

        self.login_frame = tk.Frame(self)
        self.registration_frame = tk.Frame(self)

        self.create_main_page()
        self.create_login_page()
        self.create_registration_page()

        self.show_main_page()

    def create_main_page(self):
        label = tk.Label(self.main_frame, text="Welcome to the Login and Registration System")
        label.pack(pady=10)

        login_button = tk.Button(self.main_frame, text="Login", command=self.show_login_page)
        login_button.pack(pady=5)

        registration_button = tk.Button(self.main_frame, text="Registration", command=self.show_registration_page)
        registration_button.pack(pady=5)

    def create_login_page(self):
        label = tk.Label(self.login_frame, text="Login Page")
        label.pack(pady=10)

        back_button = tk.Button(self.login_frame, text="Back", command=self.show_main_page)
        back_button.pack(pady=5)

    def create_registration_page(self):
        label = tk.Label(self.registration_frame, text="Registration Page")
        label.pack(pady=10)

        back_button = tk.Button(self.registration_frame, text="Back", command=self.show_main_page)
        back_button.pack(pady=5)

    def show_main_page(self):
        self.login_frame.pack_forget()
        self.registration_frame.pack_forget()
        self.main_frame.pack()

    def show_login_page(self):
        self.main_frame.pack_forget()
        self.registration_frame.pack_forget()
        self.login_frame.pack()

    def show_registration_page(self):
        self.main_frame.pack_forget()
        self.login_frame.pack_forget()
        self.registration_frame.pack()

if __name__ == "__main__":
    app = MainApplication()
    app.mainloop()
```

### Login Page
